Q8

```python
S3=['I', 'am', 'a', 'great', 'learner', 'I', 'am', 'going', 'to', 'have', 'an', 'awesome', 'life', 'I', 'work', 'hard', 'and', 'shall', 'be', 'rewarded', 'well', '']
removed_words= r'^(I|am|to|and|.{7,})$'
new_array = [word for word in S3 if not re.match(removed_words, word)]
length_of_new_array = len(new_array)
print("The new array is :",new_array)
print("The length of new array is :",length_of_new_array)
```

Q12

```python
import numpy as np
import matplotlib.pyplot as plt

V1 = np.random.rand(10)
sorted_vector = np.sort(V1)
plt.plot(sorted_vector,marker="o",linestyle="dashed")
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Sorted Vector")
plt.show()

squared_vector = sorted_vector ** 2
plt.plot(sorted_vector,marker="o",linestyle="dashed",color="red", label= "sorted vector")
plt.plot(squared_vector,marker="o",linestyle="dashed",color="green", label= "Squared Vector")
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparision of First and Second vector")
plt.show()
```
